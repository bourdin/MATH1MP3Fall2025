```python
%matplotlib ipympl
import matplotlib.pyplot as plt
import numpy as np
```

```python
def bisection(f, a, b, tol):
    if f(a) == 0.:
        print("The solution is: ", a)
        return a,a
    elif f(b) == 0.:
        print("The solution is: ", b)
        return b,b
    elif f(a) * f(b) >= 0:
        print("Error: the bisection method requires an initial interval [a,b] such that f(a)*f(b) <= 0, but we got f({a})*f({b}) = ", f(a)*f(b))
        return a,b

    while abs(b - a) >= tol:
        mid = (a + b)/2
        if f(mid) == 0:
            print("The solution is: ", mid)
            return mid,mid
        if f(a) * f(mid) < 0:
            b = mid
        if f(mid) * f(b) < 0:
            a = mid
    return a, b
```
