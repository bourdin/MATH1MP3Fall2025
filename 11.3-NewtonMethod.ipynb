```python
%matplotlib ipympl
import matplotlib.pyplot as plt
import numpy as np

def Newton(f,fprime,a,eps, nMax = 100):
    numIter = 0
    err = eps + 1
    while numIter < nMax and err > eps:
        numIter += 1
        c = a-f(a) / fprime(a)
        err = abs(f(c))
        a = c
        print(numIter, c, err)
    return c

def f(x):
    import numpy as np
    return np.cos(x)
def fprime(x):
    import numpy as np
    return -np.sin(x)

    

a = -np.pi/4
b = np.pi
eps = 1.e-3
Newton(f, fprime, a, eps, 10)
```

```python
X = np.linspace(a-(b-a)/2, b + (b-1)/2,101)
fig, ax = plt.subplots(figsize = (8,6))
ax.plot(X,f(X))
ax.axis((a-(b-a)/2, b + (b-1)/2,-1,1))

def NewtonWithPlot(f, fprime, a, b, eps, nMax = 100):
    ax.scatter([a,b],[f(a),f(b)])
    eps = 1.e-3
    # fig, ax = plt.subplots()
    ax.plot(X,f(X))
    numIter = 0
    err = eps + 1
    c = a
    while numIter < nMax and err > eps:
        numIter += 1
        a = c
        c = a-f(a) / fprime(a)
        err = abs(f(c))
        ax.scatter([c], [f(c)])
    ax.plot(X,f(a) + (X-a) * fprime(a))

NewtonWithPlot(f, fprime, a, b, eps, nMax = 4)
ax.grid()
```

```python
plt.close(fig)

fig2, ax2 = plt.subplots()
ax2.plot(X,f(X))
ax2.grid()
```
