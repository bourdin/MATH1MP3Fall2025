```python
%matplotlib ipympl
import matplotlib.pyplot as plt
import numpy as np

def secant(f,a,b,eps, nMax = 100):
    numIter = 0
    err = eps + 1
    while numIter < nMax and err > eps:
        numIter += 1
        c = a-f(a) * (b-a) / (f(b) - f(a))
        err = abs(f(c))
        a = b
        b = c
        print(numIter, c, err)
    return c

def f(x):
    import numpy as np
    return np.cos(x)

a = -np.pi/4
b = np.pi
eps = 1.e-3
secant(f,a,b,eps,10)
```

```python
X = np.linspace(a,b,101)
fig, ax = plt.subplots(figsize = (8,6))
ax.plot(X,f(X))


def secantWithPlot(a, b, eps, nMax = 100):
    ax.scatter([a,b],[f(a),f(b)])
    eps = 1.e-3
    # fig, ax = plt.subplots()
    ax.plot(X,f(X))
    numIter = 0
    err = eps + 1
    while numIter < nMax and err > eps:
        numIter += 1
        c = a-f(a) * (b-a) / (f(b) - f(a))
        err = abs(f(c))
        ax.scatter([c], [f(c)])
        a = b
        b = c
    ax.plot(X,f(a) + (X-a) * (f(b)-f(a))/(b-a))

secantWithPlot(a, b, eps, nMax = 4)
ax.grid()
```

```python
plt.close(fig)

fig2, ax2 = plt.subplots()
ax2.plot(X,f(X))
ax2.grid()
```
