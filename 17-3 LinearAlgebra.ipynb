# 17-3 linear algebra with scipy

## Preliminary: linear functions of a single variable.
We say that a function $f$ defined on $\mathbb{R}$ is linear if and only if for any $x,y \in \mathbb{R}$ and $\lambda, \mu \in \mathbb{R}$, we have:
$$
f(\lambda x + \mu y ) = \lambda f(x) + \mu f(y).
$$
It is easy to show that if $f$ is linear, then there exists $a \in \mathbb{R}$ such that $f(x) = ax$ for any $x$. In other words, a linear function of a single variable is fully determined by a single scalar $a$.
Furthermore, if $f$ is the linear function defined by $a$ and $g$ is the linear function defined by $b$ (*i.e.* $f(x) = ax$ and $g(y) = by$), then the function $g\circ f$ defined by $(g\circ f)(x) = g \left( f(x)\right)$ is a linear function defined by $ab$. Composition of linear functions his equivalent to multiplication of scalars. Similarly, if $f$ is the linear function with coefficient $a$ and $g$ has coefficient $b$, $h = \lambda f + \mu g$ is the linear functions with coefficient $\lambda a + \mu g$.

## Linear functions of multiple variables
The same applied to functions of several variables:
Let $f$ be a function of multiple variables $\mathbf{x} = (x_1,\dots,x_n)$. We say that $f$ is linear if and only if for any $\mathbf{x} = (x_1,\dots,x_n)$ and $\mathbf{y} = (y_1,\dots, y_n)$ and any $\lambda, \mu \in \mathbb{R}$, we have 
$$
f(\lambda \mathbf{x} + \mu \mathbf{y} ) = \lambda f(\mathbf{x}) + \mu f(\mathbf{y}).
$$

It is not hard to show that again, a linear function of $n$ variables is uniquely defined by a vector $\mathbf{a} = (a_1,\dots,a_n)$ such that
$$
f(\mathbf{x}) = a_1 x_1 + \dots + a_nx_n = \mathbf{a} \cdot \mathbf{x}\dots.
$$

Finally, again, composing of linear functions is equivalent to multiplying their coefficients:
let $f(\mathbf{x}) = a_1 x_1 + \dots + a_nx_n = a \cdot x\dots$ and $g(y) = c y$ be linear functions, then
$$
(g\circ f)(\mathbf{x}) = g\left(f(x_1,\dots,x_n)\right) = c (a_1 x_1 + \dots + a_nx_n) = c a_1 x_1 + \dots + c a_n x_n = (c \mathbf{a}) \cdot \mathbf{x}
$$
and is therefore identifiable to $c\mathbf{a}$ 


## Linear vector-valued functions
Same story...
Let $f(t)$ be a vector valued function from $\mathbb{R}$ into $\mathbb{R}^n$. We say that $f$ is linear if and only if $f(\lambda s + \mu t) = \lambda f(s) + \mu f(t)$ for any $s,t$ in $\mathbb{R}$ and any $\lambda \mu \in \mathbb{R}$, and can show that a linear function $f$ can be identified by a vector $\mathbf{a}$ such that 
$$f(t) = (a_1 t, \dots, a_n t) = t \mathbf{a}.$$

Again, all the compositions that make sense lead to multiplication (with the understanding that the multiplication of 2 vectors is in the sense of a dot product) of the coefficients.

## Putting everything together: Vector valued functions of multiple variables.
We can extend these concepts to define functions that to a independent variable in $\mathbb{R}^n$ associate an image in $\mathbb{R}^m$. For instance, we could compose $f:\ \mathbb{R}^n \mapsto \mathbb{R}$ into $g:\ \mathbb{R} \mapsto \mathbb{R}^m$. We could also think of $m$ functions $f_1,\dots, f_m$ from $\mathbb{R}^n \mapsto \mathbb{R}$, and build
$$
\mathbf{f}(\mathbf{x}) = (f_1(x_1,\dots,x_n),\dots, f_m(x_1,\dots,x_n)).
$$

As before, we could say that $\mathbf{F}$ is linear if and only $\mathbf{F}(\lambda \mathbf{x} + \mu \mathbf{y}) = \lambda \mathbf{F}(\mathbf{x}) + \mu \mathbf{F}(\mathbf{y})$, from which we would derive that each of $f_i$ must be linear, hence represented by coefficients $\mathbf{a}_i \in \mathbb{R}^n$, so that in total, $\mathbf{F}$ is represented by $m \times n$ coefficients $a_{i,j}$ with $1 \le i \le m$ and $1 \le j  \le n$. Just like $n$ reals can be thought as a vector of dimension $n$,  $m \times n$ coefficients can be thought as $m$ vectors of dimension $n$, or an $m \times n$ matrix $\mathbf{A}$:
$$
\mathbf{A} = \begin{bmatrix} a_{1,1} & \dots & a_{1,n}\\
a_{2,1} & \dots & a_{2,n}\\
\vdots &  & \vdots \\
a_{m,1}& \dots & a_{m,n}
\end{bmatrix}
$$
For instance, the coefficient matrix of the linear function 
$$
\mathbf{F}(x_1,x_2) = \begin{pmatrix} 
    x_1 & + 3 x_2 \\
    2 x_1 & - x_2 \\
    -x_1 & + 2x_2
    \end{pmatrix}
$$
is the matrix
$$
\begin{bmatrix} 
    1 & 3 \\
    2 & -1 \\
    -1 & 2
    \end{bmatrix}.
$$

## Matrix vector product.

We can define a matrix to vector product by simply writing that if $\mathbf{F}$ has coefficient matrix $\mathbf{A}$, then 
$$\mathbf{F}(\mathbf{x}) := \mathbf{A}\mathbf{x}.$$

It quickly comes that if $\mathbf{A}$ is a $m \times n$ matrix and $\mathbb{x}$ is a $n$ vector (which is basically a $n \times 1$ matrix), then $\mathbf{A}\mathbf{x}$ is a $m \times 1$ matrix such that 
$$
\left[\mathbf{A}\mathbf{x}\right]_{i,j} = \sum_{k = 1}{n} A_{i,k} x_k
$$
which is simply the dot product of the $i^\mathrm{th}$ row of $\mathbf{A}$ with $\mathbf{x}$.

## Matrix product.

Just as before, the product of 2 matrices $\mathbf{A}$ and $\mathbf{B}$ can be thought as the coefficient matrix of the linear application obtained by composing inear functions whose coefficients are $\mathbf{A}$ and $\mathbf{B}$.

This way, if $\mathbf{F}:\ \mathbb{R}^n \mapsto \mathbb{R}^p$ has for coefficients the $p\times n$ matrix $\mathbf{A}$ and $\mathbf{G}:\ \mathbb{R}^m \mapsto \mathbb{R}^p$ has for coefficients the $p\times m$ matrix $\mathbf{B}$, then $\mathbf{G} \circ \mathbf{F}$ has coefficient the $m \times n$ matrix $\mathbf{A}\mathbf{B}$ with
$$
\left[\mathbf{A}\mathbf{B}\right]_{i,j} = \sum_{k=1}^p A_{i,k} B_{k,j}.
$$

### Matrices in python
We saw that numpy has 2d arrays, which look like matrices. Addition, subtraction, and scalar multiplication work as expected but remember that numpy vectorizes all operations, so '*' is not the matrix-vector (or matrix-matrix) product:

```python
A = np.array([[1,2,3],[4,5,6]])
B = np.array([[1,-1,1],[2,2,-1]])
b = np.array([1,0,-2])
c = np.array([[1], [0], [-2]])
print("All of this is OK")
print('A:\n',A)
print('B:\n',B)
print('A+B:\n',A+B)
print('2A:\n',2*A)
print('b:\n', b)
print('c:\n', c)

print("\nThis is not OK!")
print('A*b:\n',A*b)
```

The proper way to do matrix-vector or matrix-matrix product is with the `@` or `np.matmul` functions. **This is a very common source of error.* If something looks wrong in a linear algebra calculation, check for multiplications! numpy has a `matrix` class which does products correctly but for some reason it is not widely used.


```python
Amatrix = np.matrix(A)
print(A)
print('A*b: ', A*b)
print('A*c: ', A*c)
```


Note that numpy is a bit sloppy with row vs. column vectors. `A@c` is not consistent since `A` is 2x3 and `c` is 1x3 (but `A@b` is OK since `b` is 3x1)... Or maybe the opposite if you think that numpy vectors are column vectors. Yet, numpy will gladly do both. In the case of `A@b`, what it really does is `A@(b.T)`. Whether row and column vectors of the same size are the same thing is a bit of a touchy subject. We'll leave this discussion for your linear and abstract algebra classes.

```python
print('A@c:           ', A@b)
print('np.matmul(A,c) ', np.matmul(A,c))
print('np.matmul(A,b) ', np.matmul(A,b))
```

### Example: back to Lagrange interpolation

We saw that the Lagrange interpolation of a function $f$ at $(x_0, \dots, x_{n-1})$ is a polynomial $P$ such that $P(x_i) = f(x_i)$ for $ 0 \le i \le n-1$. This can be rewritten as a linear system of equations. Let $P(x) = p_0 + p_1x + \dots + p_{n-1}x^{n-1}$, then
$$\begin{bmatrix}
1 & x_0 & x_0^2 & \dots & x_0^{n-1} \\
1 & x_1 & x_1^2 & \dots & x_1^{n-1} \\
\vdots & \vdots& \vdots& & \vdots \\
1 & x_{n-1} & x_{n-1}^2 & \dots & x_{n-1}^{n-1} \\
\end{bmatrix}
\begin{bmatrix}
p_0\\p_1\\ \vdots\\p_{n-1}
\end{bmatrix}
= 
\begin{bmatrix}
f(x_0)\\f(x_1)\\ \vdots\\f(x_{n-1})
\end{bmatrix}.
$$

```python
from numpy.polynomial import Polynomial 
def f(x):
    return np.sinc(x)

xi = np.random.uniform(size = n, low = -2, high = 2)

### Build the Lagrange polynomial by solving a linear system
M = np.empty((n,n))
for i in range(n):
    M[i,:] = [xi[i]**j for j in range(n)]
Ln = Polynomial(linalg.solve(M,f(xi)))

### Build the Lagrange polynomial directly
Ln2 = Polynomial([0])
for i in range(n):
    Lni = Polynomial.fromroots([xi[j] for j in range(n) if j != i])
    Lni = Lni / Lni(xi[i])
    Ln2 += f(xi[i]) * Lni

### Verification that teh result are the same
print(Ln.coef)
print(Ln2.coef)
print(max(Ln.coef - Ln2.coef))
```

```python
from numpy.polynomial import Polynomial 
from ipywidgets import interact

n = 15

def f(x):
    return np.sinc(x)

plt.close('all')
fig, ax = plt.subplots()
def plot(n):
    ax.cla()

    xi = np.random.uniform(size = n, low = -2, high = 2)
    ### Build the Lagrange polynomial by solving a linear system
    M = np.empty((n,n))
    for i in range(n):
        M[i,:] = [xi[i]**j for j in range(n)]
    Ln = Polynomial(linalg.solve(M,f(xi)))

    X = np.linspace(-2.5, 2.5, 201)
    ax.plot(X,f(X), label = 'f', zorder = 0)
    ax.plot(X,Ln(X), label = f'L_{n}', zorder = 1)
    ax.scatter(xi, f(xi), color = 'red', zorder = 2)
    ax.spines[['left', 'bottom']].set_position('zero')
    ax.spines[['top', 'right']].set_visible(False)
    ax.set_ylim(-1.5,1.5)
    ax.legend()

interact(plot, n = (1,20,1))
```
