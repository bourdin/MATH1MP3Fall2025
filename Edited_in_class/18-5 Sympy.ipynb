# 18-5 Advanced plotting

Instead of matplotlib, sympy can use various libraries to render plots through the Sympy Plotting Backends module. This is particularly useful as matplotlib is very slow in 3D.

### Installation

cf. https://sympy-plot-backends.readthedocs.io/en/latest/install.html

`pip install sympy_plot_backends[all]`

Then go back to your notebook, and restart the kernel (the circular arrow icon)

```python
from sympy import *
from spb import *
x = Symbol('x')
```

## Usage

Going back to our first example. At first, nothing has changed. Notice that I did not import matplotlib or run the `%matplotlib widget` magic, so I don't have zooming and other features. I could have.

```python
(sigma, mu, x) = symbols( ('sigma', 'mu', 'x') )
gaussian = exp(-(x-mu)**2/sigma**2/2)/sigma/sqrt(2*pi)
gaussian
```

```python
p = plot(gaussian.subs( {sigma:1, mu:0}))
```

Let's do the same plot with a different "backend", *i.e.* using another tool to do the actual plotting.

```python
p = plot(gaussian.subs( {sigma:1, mu:0}), backend = BB)
```

Here, we use ["Bokeh"](https://bokeh.org) to do the actual rendering and inherint all the cool stuff Bokeh can do.

Other backends for 2D plots are `MB` (matplotlib, the default option), and `PB` ([plotly](https://plotly.com))

## Additional functions offered by the Sympy Plotting Backends module.

in addition to teh functions we saw earlier, spb offers many other types of plots. 
See https://sympy-plot-backends.readthedocs.io/en/latest/modules/plot_functions/index.html for a list of plotting functions available. 

For example, here is an example of a polar coordinate plot

```python
theta = Symbol('theta')
plot_polar(3 * sin(2 * theta), (theta, 0, 2*pi), backend = BB)
```

`plot_list` is essentially the equivalent of matplotlib's plot

```python
import numpy as np
n = 25
t = Symbol('t')
exprx = 2 * sin(3*t)
expry = 3 * sin(2*t)

tt = np.linspace(0, 2*np.pi, n)
xx = 2 * np.sin(3*tt)
yy = 3 * np.sin(2*tt)

p1 = plot_parametric( exprx, expry, (t, 0, 2*pi), show = False)
p2 = plot_list(xx,yy, is_scatter = True, line_color = 'red', show = False)
p1.append(p2[0])
p1.show()

```

An additional backend (`KB`) is available in 3D. 
3D plotting with the `KB` and `BB` backends are much more fluid than with matplotlib.
`BB` does not work in 3D.

```python
(x,y) = symbols( ('x', 'y') )
plot3d(cos(x**2 + y**2), (x, -2, 2), (y, -2, 2), backend=PB)
```

```python
(x,y) = symbols( ('x', 'y') )
plot3d(cos(x**2 + y**2), (x, -2, 2), (y, -2, 2), backend=KB)
```
